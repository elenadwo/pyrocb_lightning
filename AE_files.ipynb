# AE Files: Importing Matlab

### Importing Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import cartopy
import cartopy.crs as ccrs
from pymatreader import read_mat
import os
import plotly.express as px
```

### Import Matlab Data: Read in as Python
- YYYY/MM/DD, hh:mm:ss, lat, lon, resid, nstn, Energy (J), Energy uncertainty (J), nstn_energy
- A.keys() to print out keys
- type(june28.hh)

```python
A = read_mat('data/AE20210628.mat')
june28 = pd.DataFrame(A['data'], columns = ["YYYY", "MM", "DD", "hh", "mm", "ss", "lat", "lon", "resid", "nstn", "Energy (J)", "Energy uncertainty (J)", "nstn_energy"])

B = read_mat('data/AE20210629.mat')
june29 = pd.DataFrame(B['data'], columns = ["YYYY", "MM", "DD", "hh", "mm", "ss", "lat", "lon", "resid", "nstn", "Energy (J)", "Energy uncertainty (J)", "nstn_energy"])

C = read_mat('data/AE20210630.mat')
june30 = pd.DataFrame(C['data'], columns = ["YYYY", "MM", "DD", "hh", "mm", "ss", "lat", "lon", "resid", "nstn", "Energy (J)", "Energy uncertainty (J)", "nstn_energy"])

D = read_mat('data/AE20210701.mat')
july1 = pd.DataFrame(D['data'], columns = ["YYYY", "MM", "DD", "hh", "mm", "ss", "lat", "lon", "resid", "nstn", "Energy (J)", "Energy uncertainty (J)", "nstn_energy"])

lightning = (june28, june29, june30, july1)
```

### Plotly Express Scatter: June 28

```python
points28 = (june28.loc[(june28.lon > -130) & (june28.lon < -100) & (june28.lat > 45) & (june28.lat < 65)])
fig = px.scatter(june28, x = points28['lon'], y = points28["lat"], size = points28["Energy uncertainty (J)"], color = points28["Energy (J)"], size_max = 40)

fig.show()
```

### Energy Uncertainty and Energy map: July 1

```python
# fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.BORDERS, linestyle = '-', alpha = 0.5)
ax.add_feature(cartopy.feature.COASTLINE, linestyle = '-', alpha = 0.5)
ax.add_feature(cartopy.feature.STATES, linestyle = '-', alpha = 0.5, edgecolor="black")


# Make figure larger
plt.gcf().set_size_inches(20, 10)

N = int

#print(points28.shape)
#print(points28)
points1 = (july1.loc[(july1.lon > -130) & (july1.lon < -100) & (july1.lat > 45) & (july1.lat < 65) & (july1["Energy (J)"] != 0)])
plt.scatter(x = points1["lon"], y = points1["lat"], c = points1["Energy uncertainty (J)"] ** 0.1, cmap="YlOrRd", s = np.log(points1["Energy (J)"]), edgecolors="none", marker = 'o', alpha=0.5)

plt.savefig("energy1", dpi = 200)

plt.show()
```

## Energy map: June 28 hour 6 - 8

```python
fig = plt.figure(figsize=(10,10))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.add_feature(cartopy.feature.BORDERS, linestyle = '-', alpha = 0.5)
ax.add_feature(cartopy.feature.COASTLINE, linestyle = '-', alpha = 0.5)
ax.add_feature(cartopy.feature.STATES, linestyle = '-', alpha = 0.5, edgecolor="black")


# Make figure larger
plt.gcf().set_size_inches(20, 10)
plt.xlim([-130, -100])
plt.ylim([45, 65])

N = int
june28_hour1 = june28[(june28["hh"] > 6.0) & (june28["hh"]< 8.0)]
points28 = (june28_hour1.loc[(june28_hour1.lon > -130) & (june28_hour1.lon < -100) & (june28_hour1.lat > 45) & (june28_hour1.lat < 65)])

plt.scatter(x = points28["lon"], y = points28["lat"], c = points28["Energy (J)"], cmap="hot_r", s = points28["Energy (J)"] * 0.05, edgecolors="black", marker = 'o', alpha=0.5)
plt.show()
```
